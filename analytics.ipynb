## Restoran mana yang memiliki pendapatan tertinggi dan berapa?

```python
import pandas as pd
```

```python
df = pd.read_csv('../pandas_journey/mydata/food_delivery_datasets.csv')
```

```python
df
```

```python
series_resto_status_summary = df.groupby(['resto_id', 'status']).food_price.sum()
```

```python
series_resto_status_summary
```

```python
df_total_food_price_resto = series_resto_status_summary.unstack('status')
```

```python
df_total_food_price_resto
```

```python
df_total_income_sorted = df_total_food_price_resto.sort_values(by='Completed', ascending=False)
```

```python
# remove 'Cancelled' column
df_total_income_sorted.drop('Cancelled', inplace=True, axis=1)
```
